```python
import pandas as pd

df = pd.read_csv('2020-XTern-DS.csv')
df = df[pd.to_numeric(df.Rating, errors='coerce').notnull()]
df = df[pd.to_numeric(df.Votes, errors='coerce').notnull()]

df
```

```python
df["Rating"] = df['Rating'].astype('float')
df["Votes"] = df["Votes"].astype("float")

df
```

```python
df["Average_Cost"] = df['Average_Cost'].map(lambda x: x.replace(",","."))
df["Average_Cost"] = df['Average_Cost'].map(lambda x: x.strip('$'))
df["Minimum_Order"] = df['Minimum_Order'].map(lambda x: x.strip('$'))
df["Cook_Time"] = df['Cook_Time'].map(lambda x: x.strip(' minutes'))

df
```

```python
df["Cook_Time"] = df["Cook_Time"].astype("float")
df["Average_Cost"] = df["Average_Cost"].astype("float")
df["Minimum_Order"] = df["Minimum_Order"].astype("float")

df
```

```python
cook_time = df.sort_values("Cook_Time")

cook_time
```

```python
#The average cook time is 38 minutes
df["Cook_Time"].mean()
```

```python
#The fastest cuisines are typically middle eastern, East Asian, or indian food
types_of_food = df.groupby("Cuisines")

fastest_time = types_of_food["Cook_Time"].mean()
fastest_time.nsmallest(10)
```

```python
#Orders ratings from highest to lowest
ratings = df.sort_values("Rating", ascending = False)

ratings
#We can see that the 5 most highly rated restaurants are of ID_7412, ID_4728, ID_1166, ID_6537, and ID_2051
```

```python
#dataset of all restaurants that are better than the average
top_restaurants = df[df.Rating > 3.61]

top_restaurants
```

```python
#dataset of all restaurants that are lower than average
bottom_restaurants = df[df.Rating < 3.61]

bottom_restaurants
```

```python
#The same top types of cuisine appears in the bottom restaurants as the top.
#Type of cuisine does not seem to be a factor in the rating of restaurants.
bottom_restaurants["Cuisines"].value_counts()
```

```python
#Surprisingly, lower rated restaurants have faster cook times than higher rated restaurants.
#Similarly to cuisine, cook times seem to not have an affect on top rated restaurants.
bottom_restaurants["Cook_Time"].mean()
```
